### Sentencias Print:  
Las sentencias print() son perfectas para correcciones rápidas y pequeñas. Colocando estratégicamente sentencias print dentro de tu código, puedes rastrear el flujo de ejecución, inspeccionar los valores de las variables en diferentes etapas e identificar dónde surgen las discrepancias.


```python
def calculate_average(numbers):
    total = sum(numbers)
    count = len(numbers)
    return total / count
```

Si el resultado parece incorrecto, inserta sentencias print:

```python
def calculate_average(numbers):
    print("Input numbers:", numbers)
    total = sum(numbers)
    print("Total:", total)
    count = len(numbers)
    print("Count:", count)
    average = total / count
    print("Average:", average)
    return average
print(calculate_average([2,3,4,5,6,7]))
```

### Logging:
En lugar de imprimir mensajes directamente en la consola, el módulo logging permite escribir mensajes en un archivo o una base de datos. Puedes categorizar los mensajes de log según su severidad (DEBUG, INFO, WARNING, ERROR, CRITICAL) y filtrar qué mensajes se registran según tus necesidades de depuración. Por ejemplo:

```python
import logging

# Configura el sistema de logging: guarda los mensajes en el archivo 'app.log' y establece el nivel mínimo en DEBUG (registra todo)
logging.basicConfig(filename='app.log', level=logging.DEBUG)

# debug() - Registra información detallada, útil para diagnosticar problemas durante el desarrollo
logging.debug('Este es un mensaje de depuración')

# info() - Registra mensajes informativos que confirman que el programa funciona correctamente
logging.info('Este es un mensaje informativo')

# warning() - Registra advertencias sobre situaciones inesperadas que no impiden la ejecución
logging.warning('Este es un mensaje de advertencia')

# error() - Registra errores que impiden que una función o parte del programa se ejecute correctamente
logging.error('Este es un mensaje de error')
```

El logging ofrece varias ventajas sobre las simples sentencias print:

Los mensajes de log se guardan, lo que te permite revisarlos más tarde, incluso después de que el programa haya terminado de ejecutarse.

Puedes estructurar los mensajes de log con marcas de tiempo, nombres de módulos y otros detalles relevantes, lo que facilita la localización de problemas.

Puedes ajustar los niveles de logging para controlar la cantidad de información registrada, centrándote en los detalles relevantes.

El logging es especialmente valioso para aplicaciones de larga duración, sistemas complejos y escenarios en los que necesitas rastrear eventos a lo largo del tiempo.

### Aserciones (Assertions):
Son sentencias que expresan condiciones que esperas que sean verdaderas en puntos específicos del código. Si una aserción falla, es decir, la condición no se cumple, el programa lanza una excepción, alertándote inmediatamente del problema.

Por ejemplo, en una función que calcula el área de un rectángulo, podrías verificar con una aserción que la longitud y el ancho sean positivos:

```python
def calculate_area(length, width):
    assert length > 0, "Length must be positive"
    assert width > 0, "Width must be positive"
    return length * width

print(calculate_area(2,-2))    
```

Las aserciones son herramientas poderosas para detectar errores de forma temprana en el ciclo de desarrollo.
Aseguran que la lógica interna de tu código se mantenga consistente.

Identifican situaciones en las que tu código podría recibir datos incorrectos o no válidos.

Clarifican el comportamiento esperado de tu código.
