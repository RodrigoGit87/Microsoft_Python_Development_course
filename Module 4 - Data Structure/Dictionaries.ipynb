## Diccionarios: Parejas clave-valor
Los diccionarios se encuentran entre las estructuras de datos más versátiles y utilizadas en Python. Proporcionan una potente forma de almacenar y organizar la información asociando a cada dato una clave única. Piense en ellos como en una libreta de direcciones, donde se utiliza el nombre de alguien (la clave) para buscar su información de contacto (el valor). Este mecanismo de emparejamiento clave-valor permite recuperar y manipular datos de forma eficaz.

#### Trabajar con diccionarios en Python:
Creación de un diccionario: Un diccionario se construye encerrando los pares clave-valor entre llaves {}. Cada par clave-valor está separado por dos puntos :, y los pares están separados por comas. Por ejemplo:

my_dict = {"name": "Alice", "age": 30, "city": "New York"}

En este diccionario, "nombre", "edad" y "ciudad" son claves, y "Alicia", "30" y "Nueva York" son sus valores correspondientes.

    Acceso a los valores: Para recuperar un valor del diccionario, se utiliza su clave asociada. En nuestro ejemplo, my_dict["age"] devolvería el valor 30.

    Mutabilidad y flexibilidad: Los diccionarios son estructuras dinámicas. Se pueden añadir nuevos pares clave-valor, modificar los valores asociados a las claves existentes o incluso eliminar todos los pares. Esta adaptabilidad hace que los diccionarios sean adecuados para una amplia gama de aplicaciones en las que los datos deben actualizarse o modificarse con el tiempo.

    Operaciones habituales: Los diccionarios ofrecen diversos métodos prácticos para trabajar con pares clave-valor:

        get(): Recuperar de forma segura un valor por su clave (devuelve None si la clave no existe).

        items(): Obtener todos los pares clave-valor como una lista de tuplas.

        keys(): Obtener todas las claves del diccionario.

        values(): Obtiene todos los valores del diccionario.

        update(): Fusionar otro diccionario con el existente.
