## Conjuntos: Colecciones de elementos únicos
Los conjuntos ofrecen un enfoque distinto a la organización de datos en Python. A diferencia de las listas o tuplas, los conjuntos son colecciones desordenadas, lo que significa que los elementos que las componen no tienen una secuencia o posición específica. La característica que define a los conjuntos es su insistencia en la unicidad: cada elemento sólo puede aparecer una vez dentro de un conjunto. Esta característica convierte a los conjuntos en una potente herramienta para una gran variedad de tareas.

#### Trabajar con conjuntos en Python
Creación de un conjunto: Puede crear un conjunto encerrando sus elementos entre llaves {}, separados por comas. Si intentas incluir elementos duplicados, Python los elimina automáticamente. Por ejemplo:

my_set = {1, 2, 3, 3} 

print(my_set)  # Output: {1, 2, 3}

Observa cómo el duplicado '3' se elimina al crear el conjunto.

Naturaleza desordenada: Como los conjuntos son desordenados, no puedes confiar en la posición de los elementos. Esto significa que no puedes acceder a los elementos por índice (como mi_conjunto[0]) como harías con una lista.

Operaciones clave: Los conjuntos están diseñados para realizar operaciones eficientes con elementos únicos. Puedes añadir fácilmente

add() añadir nuevos elementos a un conjunto.

remove() elementos de un conjunto.

Encontrar el union() de dos conjuntos (todos los elementos de ambos conjuntos).

Encontrar el intersection() de dos conjuntos (elementos comunes a ambos conjuntos).

Encontrar el difference() entre dos conjuntos (elementos en un conjunto pero no en el otro).

#### Aplicación real: Gustos en Redes sociales
Imagina que estás construyendo una plataforma de Redes sociales y quieres hacer un seguimiento de los usuarios únicos a los que les ha gustado un determinado POST. Un conjunto es la estructura de datos perfecta:

liked_by = {12345, 67890, 98765} # User IDs

Cada ID de usuario es único dentro de este conjunto, lo que evita duplicados y garantiza un recuento preciso de las personas que han interactuado con la publicación.

Los conjuntos son la solución cuando se trata de elementos únicos o cuando se necesita realizar operaciones de conjunto como encontrar elementos comunes o diferencias.

```python
# Los conjuntos son mutables por defecto (se pueden añadir y eliminar elementos), 
# pero hay un tipo especial de conjunto llamado frozenset que es inmutable.
# Si no ha seleccionado intencionadamente frozenset, se trata de un conjunto mutable normal.

# ¿Necesitas quitar duplicados de una lista? Conviértala en un conjunto y Python se encargará automáticamente de ello
# Ejemplo práctico: Conjuntos en acción

numbers = [1, 2, 2, 3, 4, 4, 5]
unique_numbers = set(numbers)    # Elimina duplicados
print(unique_numbers)
```
