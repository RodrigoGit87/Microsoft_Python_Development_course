## Listas: Tus contenedores de datos versátiles
Las listas se encuentran entre las estructuras de datos más versátiles de Python. Dentro de una sola lista, puedes almacenar una variada colección de elementos, ya sean números, cadenas, otras listas, o incluso una mezcla de diferentes tipos de datos. Esta adaptabilidad hace de las listas un bloque fundamental para muchos programas Python.

#### Cómo trabajar con Listas:
Creación de una lista: Construir una lista es tan sencillo como encerrar sus elementos entre corchetes y separarlos con comas. Por ejemplo:

my_list = [1, 2.1, "hello", [3, 4]]

En este ejemplo, mi_lista contiene un entero (1), un número en coma flotante (2,1), una cadena ("hola"), e incluso otra lista ([3, 4]).

    Acceso a los elementos (indexación): Puedes localizar y recuperar elementos específicos de una lista utilizando su índice, que representa su posición en la lista. Ten en cuenta que Python, como muchos lenguajes de programación, empieza a contar desde 0. Así, mi_lista[0] te daría el primer elemento (1), mi_lista[1] el segundo (2.1), y así sucesivamente.

    Extracción de subconjuntos: Las listas permiten tomar una porción de la lista. Especifique los índices inicial y final (separados por dos puntos) para definir el intervalo que desee. Tenga en cuenta que el índice final no es inclusivo. Por ejemplo, mi_lista[1:3] extraería los elementos 1 y 2 (el segundo y tercer elemento), dando como resultado [2.1, "hola"].

    Extender listas (extend): Si quieres añadir múltiples elementos de otra lista a tu lista existente, puedes utilizar el método `extend()`. Esto es más eficiente que usar `append()` repetidamente. Por ejemplo, puedes añadir el contenido de la lista nuevas_tareas al final de la lista de tareas:

tasks = ["Take a nap"]

new_tasks = ["Go for a walk", "Read a book"]

tasks.extend(new_tasks)

    Modificación de listas (mutabilidad): Una característica definitoria de las listas es su mutabilidad. No está limitado a los contenidos iniciales de una lista; puede añadir dinámicamente nuevos elementos utilizando append(), insertar elementos en posiciones específicas con insert(), eliminar elementos por valor con remove(), o por posición con pop(). También puede reorganizar el orden de los elementos mediante sort() e reverse(). Esta flexibilidad le permite manipular listas para adaptarlas a las necesidades cambiantes de su programa.

Aplicación práctica: La lista de tareas
Imagina que estás desarrollando una aplicación de listas de tareas. Una lista de Python es la forma más natural de almacenar tus tareas:

tasks = ["Buy groceries", "Finish report", "Call Mom"]

Podrías añadir fácilmente una nueva tarea (tasks.append("Programar cita con el dentista")), marcar una tarea como completada eliminándola (tasks.remove("Terminar informe")), o incluso ordenar tus tareas alfabéticamente (tasks.sort()).

Las listas son increíblemente flexibles y una buena opción cuando se necesita almacenar una colección de elementos que es posible que desee modificar más adelante

```python
my_list = [1, 2.1, "hello", [3, 4]]
print(my_list[0])# 1

parcial_list = my_list[1:3] #[2.1, 'hello']
print(parcial_list)

first_elements = my_list[:2]
print(first_elements)

tasks = ["Buy groceries", "Finish report", "Call Mom"]
tasks.append("Coding 24/7 !")
print(tasks)

#tasks.remove("Finish report") or remove by index;
del tasks[1]
print(tasks)
```
