### Tuplas: Colecciones de datos inmutables
Las tuplas son como las listas, pero con una diferencia clave: son inmutables. Una vez creada una tupla, no se puede modificar su contenido. Esto las hace ideales para almacenar datos que no deben modificarse. No podrás añadir nuevos elementos, eliminar los existentes ni modificar sus valores. Aunque esto pueda parecer restrictivo a primera vista, es una característica muy poderosa cuando necesitas salvaguardar los datos de cambios accidentales o involuntarios.

#### Cómo trabajar con tuplas:
Creación de una tupla: La creación de una tupla es similar a la creación de una lista, pero en lugar de corchetes, se utilizan paréntesis para encerrar los elementos. Por ejemplo:

my_tuple = (10, 20, "python")

Aquí, mi_tupla contiene tres elementos: los enteros 10 y 20, y la cadena "python"

    Acceso a los elementos (indexación y segmentación): Al igual que las listas, puedes acceder a los elementos individuales de una tupla utilizando su índice, y puedes extraer un subconjunto de elementos utilizando la segmentación. La sintaxis es idéntica. Por ejemplo, mi_tupla[2] devolvería la cadena "python", y mi_tupla[0:2] los dos primeros elementos (10, 20).

    La inmutabilidad en acción: La inmutabilidad de las tuplas significa que una vez que hayas definido mi_tupla como se muestra arriba, no podrás hacer algo como mi_tupla[1] = 30 o mi_tupla.append("nuevo elemento"). Python emitirá un error si intentas modificar el contenido de la tupla.

#### Aplicación en el mundo real: Coordenadas geográficas
Imaginemos que trabajamos con coordenadas geográficas (latitud y longitud). Una tupla es la estructura de datos perfecta para representar estas coordenadas:

coordinates = (37.7749, -122.4194)  # Latitude and longitude of San Francisco

La inmutabilidad de la tupla garantiza que estas coordenadas permanezcan fijas y precisas, lo que es crucial para aplicaciones como la cartografía o los sistemas de navegación.

Las tuplas son muy útiles cuando quieres asegurarte de que los datos permanecen constantes a lo largo de la ejecución de tu programa. Su inmutabilidad garantiza que, una vez creada una tupla, su contenido no se alterará ni accidental ni intencionadamente, lo que proporciona una capa de integridad de los datos que puede ser crucial en determinadas aplicaciones.
