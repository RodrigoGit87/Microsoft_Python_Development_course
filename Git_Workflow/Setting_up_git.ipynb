### ⚙️ Configurando Git  
Después de la instalación, es crucial configurar Git con tu información personal.

```bash
git config --global user.name "Tu Nombre"
```

```bash
git config --global user.email "tu_email@ejemplo.com"
```

---

### 📂 Creando un Repositorio Git  
Ahora que Git está instalado y configurado, vamos a crear un repositorio Git para rastrear los cambios en tu proyecto. Navega al directorio de tu proyecto en tu terminal o consola y ejecuta el siguiente comando:  

```bash
git init
```

---

### 📦 Preparando y confirmando cambios  

**Preparar cambios (Staging):**  
```bash
git add <nombre_del_archivo>
```

**Confirmar cambios (Committing):**  
```bash
git commit -m "Tu mensaje de commit"
```

---

### 🌿 Ramas y fusiones (Branching & merging)  
Una rama es esencialmente una versión paralela de tu proyecto, que te permite trabajar en nuevas características o en la corrección de errores sin afectar el código base principal.

**Creando una rama:**  
```bash
git branch <nombre_de_la_rama>
```

**Fusionando ramas:**  
Una vez que hayas completado tu trabajo en una rama, puedes fusionarla de vuelta a la rama principal (generalmente llamada `master` o `main`). Primero, cambia a la rama principal y luego ejecuta el siguiente comando:

```bash
git merge <nombre_de_la_rama>
```

---

### 🌐 Repositorios remotos  

**Clonar un repositorio:**  
Para obtener una copia local de un repositorio remoto, usa el siguiente comando:  
```bash
git clone <url_del_repositorio>
```

**Descargar cambios (Pulling):**  
Para obtener y fusionar los cambios desde el repositorio remoto hacia tu repositorio local, usa el siguiente comando:  
```bash
git pull
```
