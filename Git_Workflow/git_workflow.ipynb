### 🔄 Un flujo de trabajo típico en Git

**1. Inicializar un repositorio (`git init`):**  
El primer paso es crear un repositorio Git para tu proyecto usando el comando `git init`. Esto configura la carpeta oculta `.git` que rastreará todos tus cambios.

**2. Hacer cambios y prepararlos (`git add`):**  
A medida que trabajas en tu código, harás cambios en los archivos de tu proyecto. Antes de poder confirmar estos cambios, necesitas prepararlos (stage) usando el comando `git add`. Esto le dice a Git qué cambios quieres incluir en tu próximo commit.

**3. Confirmar tus cambios (`git commit`):**  
Una vez que has preparado tus cambios, puedes crear una confirmación usando el comando `git commit -m "Tu mensaje de commit"`. El mensaje del commit debe ser una descripción concisa de los cambios que hiciste.

**4. Crear ramas (opcional) (`git branch`):**  
Si estás trabajando en una nueva característica o corrección de errores, a menudo es una buena idea crear una nueva rama usando el comando `git branch`. Esto te permite trabajar en tus cambios de forma aislada sin afectar el código base principal.

**5. Cambiar de rama (opcional) (`git checkout` o `git switch`):**  
Para empezar a trabajar en tu nueva rama, necesitas cambiarte a ella usando el comando `git checkout` (o `git switch`). Esto actualiza tu directorio de trabajo para reflejar el estado de la rama a la que cambiaste.

**6. Hacer más cambios y confirmar:**  
Continúa haciendo cambios y confirmándolos a medida que trabajas en tu nueva característica o corrección.

**7. Fusionar tu rama (opcional) (`git merge`):**  
Una vez que hayas completado tu trabajo en la rama, puedes fusionarla de vuelta a la rama principal (generalmente `main` o `master`) usando el comando `git merge`. Esto integra tus cambios en la línea principal de desarrollo.

**8. Subir tus cambios (`git push`):**  
Finalmente, puedes enviar (push) tus commits locales a un repositorio remoto (como GitHub) usando el comando `git push`. Esto hace que tus cambios estén disponibles para otros que colaboran en el proyecto.

**9. Traer cambios de otros (`git pull`):**  
Para mantenerte al día con los últimos cambios realizados por tus colaboradores, puedes extraer sus commits del repositorio remoto usando el comando `git pull`.
