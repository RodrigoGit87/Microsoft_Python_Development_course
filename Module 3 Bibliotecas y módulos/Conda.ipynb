## ¿Qué es Conda y cómo funciona?
Conda es un sistema de gestión de paquetes y entornos de código abierto y multiplataforma. Mientras que pip se centra principalmente en la gestión de paquetes de Python, Conda amplía sus capacidades para gestionar paquetes y dependencias en varios lenguajes de programación y plataformas (como R, Java, etc.). Conda consigue esto creando entornos aislados donde puedes instalar versiones específicas de paquetes sin interferir con otros proyectos o con la instalación base de Python de tu sistema. Es como tener múltiples cajas de herramientas autónomas para diferentes proyectos.

#### Comandos importantes de Conda
conda create: Crea un nuevo entorno Conda.

conda activate: Activa un entorno Conda específico.

conda deactivate: Desactiva el entorno actualmente activo.

conda install: Instala paquetes en el entorno activo.

conda list: Lista todos los paquetes instalados en el entorno activo.

conda search: Busca paquetes disponibles en los repositorios de Conda.

conda search <keyword>: Busca en los repositorios de Conda paquetes relacionados con la palabra clave dada.

conda update: Actualiza un paquete a su última versión.

conda remove: Desinstala un paquete del entorno activo.

conda env list: Lista todos los entornos Conda en su sistema.

#### Gestión de paquetes con Conda
conda install <package-name>: Instala el paquete especificado en tu entorno Conda activo.

conda list <package-name>: Muestra información sobre el paquete especificado, incluyendo su versión.

conda update <package-name>: Actualiza el paquete especificado a su última versión disponible.

conda install <package-name>=<version>: Instala una versión específica del paquete especificado.

conda remove <package-name>: Desinstala el paquete especificado de su entorno activo
