### ¿Qué son realmente las pruebas unitarias (Unit testing)?
Es una forma sistemática de dividir tu código en componentes fundamentales y comprobables; a estos los llamamos "unidades" (units). Las unidades suelen ser funciones (fragmentos de código que realizan una tarea específica) o clases (planos para crear objetos). La clave aquí es que cada unidad debe ser relativamente autónoma y estar enfocada.

Una vez que hemos identificado nuestras "unidades", creamos "casos de prueba" (test cases). Estos son esencialmente programas en miniatura, escritos específicamente para poner a prueba cada unidad. Un caso de prueba provee diferentes entradas a la unidad, observa cómo reacciona y luego verifica que la salida coincida con nuestras expectativas. La idea es simular una amplia gama de escenarios, incluyendo tanto casos de uso típicos como casos límite.

### Key terminology
- Aserciones (Assertions)
 Son declaraciones que verifican si tu código se está comportando como se espera. Digamos que tienes una función llamada sumar_numeros(2, 3). Una aserción comprobaría si el resultado de esta función es, de hecho, 5. Si es así, la aserción "pasa" (passes). Si no lo es, "falla" (fails), alertándote de un posible error.

- Casos de prueba (Test cases)
Un caso de prueba es un conjunto de estas aserciones. Cada aserción en un caso de prueba se enfoca en un aspecto específico del comportamiento de la función, asegurando que maneje correctamente diversas entradas y situaciones.

- Suite de pruebas (Test suite) 
Engloba múltiples casos de prueba, cada uno de ellos dirigido a diferentes unidades dentro de un módulo más grande, o incluso en la totalidad de tu base de código.
