### Que es pytest?
pytest es un marco de pruebas (framework) que facilita la escritura de pruebas pequeñas y legibles, y permite escalarlas para soportar pruebas funcionales complejas para tus aplicaciones y bibliotecas.

Para ejecutar el test en un archivo. Abrir la terminal y escribir:

 **pytest**

Buscará por todas las subcarpetas del proyecto y  ejecutará cualquier archivo que empieze por test_

Si queremos ejecutar el pytest de un archivo especifico, abrir la terminal, asegurarnos de que estamos en la ruta completa y escribir 

**pytest archivo.py**

O darle la ruta completa,entrecomillada, estemos donde estemos en la terminal. Ejemplo:

**pytest "C:\Curso Python\Module 6 - Testing basisc & Version Control\test_calculator.py"**


### Conceptos y características clave de pytest

#### 1. Funciones de Prueba (Test Functions)
Son el corazón de pytest. Consisten en funciones simples de Python a las que se les añade el prefijo `test_` en el nombre para que pytest las reconozca automáticamente.

Dentro de estas funciones usas la palabra clave `assert` para comprobar que tu código da el resultado esperado. Si el `assert` falla (ej. esperabas 5 y dio 4), pytest te avisa de que hay un error que requiere tu atención.

#### 2. Fixtures ("Preparadores" del entorno)
Piensa en los fixtures como el equipo de tramoyistas de un teatro. Se encargan de preparar el escenario antes de realizar la prueba y de recogerlo todo al terminar. Garantizan que cada prueba se ejecute en un entorno limpio y aislado.

Por ejemplo: te sirven para crear una pequeña base de datos temporal o simular que llamas a una API externa sin conectarte a internet de verdad. Fomentan enormemente la reutilización de código en tus tests y evitan que un test ensucie los datos de otro test

Creas una función, le pones este decorador, y luego pytest inyectará el resultado de esa función en cualquier test que lo pida como argumento.
```python
import pytest

# 1. Defines el fixture
@pytest.fixture
def datos_de_prueba():
    lista = [1, 2, 3, 4, 5]
    return lista

# 2. Lo usas en tu test simplemente poniéndolo como parámetro
def test_suma_lista(datos_de_prueba):
    assert sum(datos_de_prueba) == 15
```

#### 3. Parametrización (Parametrization)
Básicamente le dice a pytest: "¡No pruebes solo un caso, pruébalos todos!".

Permite ejecutar **la misma función de prueba varias veces**, pero pasándole diferentes datos de entrada en cada ejecución. Es ideal para comprobar de forma limpia cómo se comporta tu código ante muchos escenarios diferentes sin tener que escribir funciones de test redundantes.

#### 4. Plugins (Extensiones)
El verdadero superpoder de pytest es que es muy ampliable. La comunidad ha desarrollado cientos de plugins para casi cualquier necesidad a la hora de hacer testing.

Da igual si estás probando aplicaciones web, bases de datos o código asíncrono; casi seguro que hay un plugin que encaja y te agiliza el trabajo.

### Marcadores comunes de pytest (pytest markers)
Los marcadores (`markers`) en pytest se usan para clasificar las pruebas o modificar cómo se ejecutan. Aquí tienes los más comunes:

#### 1. `@pytest.mark.skip`
Omite la ejecución de ese test incondicionalmente. Útil si la prueba está rota o la función aún no está terminada.
```python
import pytest

@pytest.mark.skip(reason="Función aún no implementada")
def test_func_futura():
    assert False
```

#### 2. `@pytest.mark.skipif`
Omite el test **solamente si** se cumple una condición específica (por ejemplo, estás en el sistema operativo incorrecto o no hay conexión a internet).
```python
import pytest
import sys

@pytest.mark.skipif(sys.platform == 'win32', reason="No se ejecuta en Windows")
def test_linux_only():
    assert True
```

#### 3. `@pytest.mark.xfail`
Le indica a pytest que **esperas** que la prueba falle (ej. por un error conocido del sistema). Si efectivamente falla, pytest no lo tratará como un error fatal, sino como un fallo esperado (`xfailed`).
```python
import pytest

@pytest.mark.xfail(reason="Hay un bug conocido en esta función que está pendiente de arreglo")
def test_bug_conocido():
    assert 1 == 0
```

#### 4. `@pytest.mark.parametrize`
¡Uno de los más potentes! Permite ejecutar **el mismo test** varias veces pasando diferentes datos de entrada. En lugar de escribir 5 funciones casi idénticas copiando y pegando, simplemente parametrizas una sola.
```python
import pytest

@pytest.mark.parametrize("a, b, resultado_esperado", [
    (1, 2, 3),   # Primera ejecución
    (5, 5, 10),  # Segunda ejecución
    (10, 20, 30) # Tercera ejecución
])
def test_sumas_multiples(a, b, resultado_esperado):
    assert a + b == resultado_esperado
```

### Convenciones de nombre
Los nombres descriptivos para tus funciones de prueba y archivos son esenciales para una organización eficiente de las pruebas. Un nombre claro y específico, como `test_calculate_total_price`, te indica inmediatamente el propósito de esa prueba.

### Organizacion

La modularidad en las pruebas significa organizar tus pruebas en unidades más pequeñas e independientes. Cada prueba se centra en una parte específica de tu código, asegurando que funcione correctamente por sí sola. Esto hace que sea más fácil identificar y solucionar problemas, similar a diagnosticar una máquina revisando cada componente individualmente. También te permite cambiar o actualizar partes de tu código sin afectar a toda la suite de pruebas, al igual que reemplazar una pieza en una máquina sin desmontarla por completo.

Este código Python implementa funcionalidades básicas de comercio electrónico como añadir productos a un carrito, calcular totales con impuestos y descuentos, y simular un proceso de pago, junto con pruebas unitarias para asegurar su correcto funcionamiento:

```python
import pytest

class Product:
    def __init__(self, name, price, quantity=1):
        self.name = name
        self.price = price
        self.quantity = quantity

class Cart:
    def __init__(self):
        self.items = []

    def add_to_cart(self, product):
        self.items.append(product)

    def calculate_total_price(self, tax_rate=0.0, shipping_fee=0.0):
        subtotal = sum(item.price * item.quantity for item in self.items)
        tax = subtotal * tax_rate
        total = subtotal + tax + shipping_fee
        return total

    def apply_discount(self, discount_percentage):
        for item in self.items:
            item.price *= (1 - discount_percentage / 100)

    def checkout(self, payment_gateway):
        total_price = self.calculate_total_price()
        # In a real application, you'd interact with the payment_gateway here
        if payment_gateway.process_payment(total_price):
            # Place the order (logic not shown here)
            return True
        else:
            return False

class TestProductFunctions:
    def test_add_to_cart_successful(self):
        cart = Cart()
        product = Product("Widget", 10.0)
        cart.add_to_cart(product)
        assert len(cart.items) == 1
        assert cart.items[0] == product

    def test_calculate_total_price_accuracy(self):
        cart = Cart()
        cart.add_to_cart(Product("Widget A", 15.0))
        cart.add_to_cart(Product("Widget B", 20.0, 2))  # 2 of these
        total = cart.calculate_total_price(tax_rate=0.07, shipping_fee=5.0)
        expected_total = (15.0 + 20.0 * 2) * 1.07 + 5.0
        assert total == expected_total

    def test_apply_discount_correctly(self):
        cart = Cart()
        cart.add_to_cart(Product("Discounted Item", 50.0))
        cart.apply_discount(20)  # 20% discount
        assert cart.items[0].price == 40

    def test_checkout_successful_payment(self, mocker):
        # Mocking a payment gateway (would be a real integration in a production app)
        mock_payment_gateway = mocker.Mock()
        mock_payment_gateway.proces_payment.return_value = True

        cart = Cart()
        cart.add_to_cart(Product("Something", 100.0))
        result = cart.checkout(mock_payment_gateway)
        assert result is True # Checkout should be successful
```
