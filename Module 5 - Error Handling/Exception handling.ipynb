## El bloque try-except-else-finally
El mecanismo de manejo de excepciones en Python se basa en cuatro palabras clave fundamentales.

#### try: 
Aquí es donde colocas el código que podría causar un error. Por ejemplo, si estás intentando abrir y leer un archivo, pones ese código dentro de un bloque try. Básicamente estás diciendo: 'Intenta hacer esto, pero prepárate por si algo sale mal.'

#### except:
 El bloque except es tu plan de contingencia. Especificas el tipo de excepción que esperas capturar y defines las acciones que tu código debe tomar si esa excepción específica ocurre. Por ejemplo, si esperas un FileNotFoundError, puedes escribir un bloque except FileNotFoundError que imprima un mensaje de error y quizás le pida al usuario una ruta de archivo diferente.

#### else:
 El bloque else se ejecuta únicamente si el código dentro del bloque try **no** lanzó ninguna excepción. Es útil para colocar código que solo debe ejecutarse cuando todo ha salido bien, separándolo del código que podría fallar. Por ejemplo, si abres un archivo correctamente en el try, puedes procesar su contenido en el else.

#### finally:
 Este bloque es la limpieza opcional de tu código, asegurando que las acciones críticas se ejecuten sin importar si ocurren excepciones o no. El código dentro del bloque finally está garantizado a ejecutarse, lo que lo hace esencial para liberar recursos adquiridos en el bloque try. Se cubrirá en profundidad en materiales posteriores cuando se manejen recursos como conexiones a bases de datos.

Aquí hay un ejemplo de código que incorpora la estructura try-except para lectura de archivos, dejando de lado else y finally por ahora.

````python
file_name = "sample.txt"
try:
    with open(file_name, 'r') as file:
        contents = file.read()
        print(contents)
except FileNotFoundError:
    print("Error: File not found -", file_name)
````

## Elegir la excepción correcta
Seleccionar la excepción apropiada para tu bloque except es crucial:
- **ZeroDivisionError**: Esta excepción se lanza cuando intentas dividir un número entre cero.
- **FileNotFoundError**: Intentar acceder a un archivo que no existe dispara esta excepción (como se muestra en el ejemplo anterior).
- **IndexError**: Esta excepción surge cuando intentas acceder a un elemento en una secuencia (como una lista o una cadena de texto) usando un índice inválido. Por ejemplo, intentar acceder al décimo elemento en una lista que solo tiene 5 elementos lanzaría un IndexError.
- **KeyError**: Si intentas acceder a una clave que no existe en un diccionario, obtendrás un KeyError.

## Registro de eventos (Logging)
El logging es la práctica de registrar detalles sobre las excepciones y otros eventos en tu código. Esto crea un historial que es invaluable para depurar y entender cómo se comporta tu aplicación en el mundo real.
```python
import logging

try:
    # Tu código propenso a errores aquí
except Exception as e:
    logging.error(f"An error occurred: {e}")  # Registra la excepción con detalles
```
El logging proporciona contexto, haciendo más fácil identificar la causa raíz de los problemas.


## Lanzar excepciones personalizadas
A veces te encontrarás con errores que no encajan en ninguna de las excepciones existentes. En esos casos, puedes crear excepciones personalizadas:

```python
class InvalidCredentialsError(Exception):
    pass

# ... más adelante en tu código ...
if not valid_credentials(username, password):
    raise InvalidCredentialsError("Incorrect username or password")
```
